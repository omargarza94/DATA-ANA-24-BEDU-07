```python
import re
import pandas as pd

df = pd.read_csv('tmp\Ejemplo_01_03_Amazon_Sales_Dataset.csv')

df.head()

def find_gmail_emails(text):
    pattern = r'@gmail\.com'
    result = re.findall(pattern, text)
    return bool(result)

df_gmail = df[df['Customer Email'].apply(find_gmail_emails)]

df_gmail.head()
```
